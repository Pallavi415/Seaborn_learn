```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Categorical Data Plots
--->---->--->--->---->--->
There are a few main plot types for this:

--> boxplot
--> violinplot
--> stripplot
--> swarmplot
--> barplot
--> countplot

```python
df = sns.load_dataset("tips")
df
```

COUNT PLOT
-->-->-->-->-->-->
A Count Plot shows the number of occurrences of each category in a dataset.  
It helps compare the frequency of different categorical values.

```python
sns.countplot(df['sex'])
```

```python
sns.countplot(x=df['sex'], hue = df['smoker'])#hue is uesd to divide based on category
```

BAR PLOT
-->--->-->
A Bar Plot compares the average (or other statistical value) of a numerical variable across different categories.  
It helps compare values between different groups.

```python
sns.barplot(x=df['sex'], y = df['total_bill'])
#bar plot is created using x and y axis, at x--> categorical data is gone or used and at y - axis --> quantitative data is used 
```

you can change the estimator object to your own function, that converts a vector to a scalar Mean (default): np.mean Median: np.median Sum: np.sum Count: len (to count the number of observations) Standard Deviation: np.std Minimum: np.min Maximum: np.max

```python
import numpy as np 
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
```

BOX PLOT
--->--->
A Box Plot visualizes the distribution of numerical data using the **five-number summary**.  
It also helps identify the spread of the data and detect outliers.
---->----->--->
Five-Number Summary
--> Minimum
--> First Quartile (Q1)
--> Median (Q2)
--> Third Quartile (Q3)
--> Maximum

```python
sns.boxplot(x = 'tip',y='day',data = df)
```

BOXPLOT AND VIOLINPLOT

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.violinplot(x = 'day',y = 'total_bill',data = df)
```

STRIPPLOT
A **Strip Plot** displays individual data points across different categories.  
It helps visualize the distribution and spread of the data.
Strip Plot = Every data point is shown.    

SWARMPLOT -->A **Swarm Plot** displays individual data points across categories without overlapping.  
It helps show the exact distribution and density of the data.
    -->Swarm Plot = Strip Plot + No Overlapping.

```python
sns.stripplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```

```python

sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df)
```

Plot Memory Trick
Count Plot	-->  Count categories
Bar Plot	-->  Average values
Box Plot	-->  5-number summary + Outliers
Violin Plot	-->  Box Plot + KDE
Strip Plot	-->  Every data point
Swarm Plot	-->  Strip Plot + No Overlap
