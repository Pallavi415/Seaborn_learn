HEAT MAP

## Heatmap
A **Heatmap** visualizes data in a matrix using different colors to represent values.  
It helps identify patterns, correlations, and high or low values quickly.

```python
flights = sns.load_dataset('flights')
flights
```

```python

tips = sns.load_dataset("tips")
tips
```

heatmap is used to show corelation

```python
tipscorr = tips[['total_bill','size']]
tipscorr
```

```python
tipscorr.corr()#corealation :-always between 1 and 0
```

```python
sns.heatmap(tipscorr.corr(),annot = True)
```

CLUSTER MAP

```python
sns.clustermap(tipscorr.corr())
```

pivot table Heat map

```python

pvtflight = flights.pivot_table(values = 'passengers',index = 'month',columns = 'year')
```
