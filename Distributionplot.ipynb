- Identify variables by **what they represent**, not by their data type (`int`, `float`, `object`).
- **Label/Identifier** → Qualitative (e.g., `sex`, `day`, `Employee_ID`).
- **Count** → Quantitative (Discrete) (e.g., `size`, number of students).
- **Measurement** → Quantitative (Continuous) (e.g., `total_bill`, `tip`).
- **Why is `size` often treated as categorical?** Although `size` is a **discrete quantitative** variable, it has only a few unique values (1–6), so Seaborn treats each value as a separate group/category for comparison in plots.

```python
sns.displot(df['total_bill'])
```

## Histogram

A **histogram** shows how the values of a **numerical variable** are distributed.  
It groups the data into **bins** and displays how many values fall into each bin.

```python
sns.displot(df['total_bill'],bins = 6)
```

```python
sns.histplot(df['total_bill'],bins = 100)
```

## Kernel Density Estimation (KDE)

A **Kernel Density Estimation (KDE)** plot is a **smooth version of a histogram** that shows how data is distributed.  
It helps you understand the overall pattern of the data without using bars.

```python
plt.subplot(2,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(2,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

JOINT PLOT

```python
### Joint Plot
#A **Joint Plot** is used to visualize the **relationship between two numerical variables** along with their individual distributions.  
#It combines a **scatter plot** in the center with **histograms or KDE plots**
#on the sides.
sns.jointplot(x='total_bill', y ='tip',data=df,kind = 'scatter')
```
